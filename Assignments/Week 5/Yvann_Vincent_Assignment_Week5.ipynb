# IMPORT NECESSARY LIBS

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./diamonds.csv")
```

```python
# No undefined values so no data cleaning needed in that regard

data.isna().sum()
```

# PAIRPLOT

```python
cols = ["price", "carat", "x", "y", "z", "cut"]
sns.pairplot(data=data[cols], hue='cut')
```

# HEATMAP

```python
cols = ["carat", "depth", "table", "price", "x", "y", "z", "cut"]

fair_cut = data[cols][data["cut"] == "Fair"]
ideal_cut = data[cols][data["cut"] == "Ideal"]
```

```python
_, ax = plt.subplots(
    1,2,
    figsize = (20,10)
)

fair_heatmap = sns.heatmap(
    data = fair_cut.corr(),
    annot = True,
    cmap = "rocket_r",
    ax = ax[0]
).set(
    title = "Corr Coeff for cut=Fair"
)

ideal_heatmap = sns.heatmap(
    data = ideal_cut.corr(),
    annot = True,
    cmap = "rocket_r",
    ax = ax[1]
).set(
    title = "Corr Coeff for cut=Ideal"
)
```

# BOX PLOTS

```python
_, ax = plt.subplots(figsize = (10,10))

cols = ["carat", "cut", "color"]

sns.boxplot(data = data[cols], x = "cut", y = "carat", hue = "color")
```

```python
_, ax = plt.subplots(figsize = (10,10))

cols = ["price", "cut", "color"]

sns.boxplot(data = data[cols], x = "cut", y = "price", hue = "color")
```

# SCATTER PLOT

```python
_, ax = plt.subplots(figsize = (10,10))

cols = ["price", "carat", "cut"]

sns.scatterplot(data = data[cols], x = "carat", y = "price", hue = "cut")
```

# GRID OF SCATTERPLOTS

```python
grid = sns.FacetGrid(data, col="cut", hue="cut")
grid.map(sns.scatterplot, "carat", "price", alpha=.8)
grid.add_legend()
```
