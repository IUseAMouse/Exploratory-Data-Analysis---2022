```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("./salaries.csv")
```

```python
data.columns
```

```python
fig, ax = plt.subplots(figsize=(10,10))
data_per_xp = data[data["employee_residence"] == "US"].groupby(["experience_level"]).mean()

sns.barplot(data=data_per_xp, y="salary_in_usd", x=data_per_xp.index, ax=ax).set(
    title="Average salary in $ by experience level"
)

fig.savefig("plot.jpg")
```

```python
genre = [
    "Action",
    "Adventure",
    "Magic",
    "Fantasy",
    "Romance",
    "Comedy",
    "Drama"
]

count = [
    586,
    483,
    232,
    84,
    326,
    127,
    184
]

raph_data = pd.DataFrame(columns=("genre","count"))
```

```python
fig, ax = plt.subplots()

raph_data["genre"] = genre
raph_data["count"] = count 

sns.barplot(data=raph_data, x="genre", y="count", ax=ax).set(
    title = "Number of anime per traditional genre in the dataset"
)

fig.savefig("raph.jpg")
```
